**알아야 할 상식:** 

* 위 흑백 사진은 브라우저가 사람 눈에 보기 좋게 처리하여  보여주는 것이다. 
    하지만 컴퓨터에 그림이 저장될 때는 픽셀(pixel)들의 수치 데이터로 저장된다.
    pixel의 어원은 그림을 나타내는 picture의 줄임말 'pix'와 
    구성요소를 의미하는 element의 줄임날 'el'을 합쳐서 만들어졌다. 
    픽셀을 우리말로 화소(畵素, 그림의 원소)라고 부르는 이유가 이때문이다. 


* 하나의 픽셀은 보통 세 개 또는 네 개의 값들로 구성된다. 
    * 세 개의 값으로 구성된 경우: 
        세 개의 값은 각각 Red(빨강색), Green(녹색), Blue(파랑색) 정보에 해당하는
        숫자들이다. 
        즉, RGB 색 모델을 구성한다.

    * 네 개의 값으로 구성된 경우: 
        네 번째 항목은 색의 투명도를 나타낸다. 

#### 사진의 데이터 확인하기
이제 위 사진의 수치 데이터를 `plt.imread()` 함수를 이용하여 확인해 보자.

```python
img_hopper = plt.imread('../../images/two_D_data/grasshopper.png')
img_hopper
```

`img_hopper` 변수에 저장된 어레이의 모양(shape)은 다음과 같다.

```python
img_hopper.shape
```

즉, 3차원 어레이이다. 3차원 어레이는 일반적으로 상상하기가 매우 어렵다.
하지만 사진 데이터의 경우에는 다음처럼 이해하면 매우 쉽다.

* 먼저, 해당 사진에 세로를 균등하게 329개로 쪼개고, 가로를 500개로 쪼갠다.
* 그러면 `329 x 500` 크기의 격자 모양으로 사진을 쪼개었다고 생각할 수 있다. 
* (격자 사진 넣기)
* 이제 각각의 격자 칸이 픽셀에 해당하며, 픽셀 정보는 길이가 4인 어레이로 표현되어 있다.

#### 수치 데이터에서 사진 재생하기

사진의 수치 데이터를 이용하여 사진을 재생하려면 `plt.imshow()` 함수를 이용하면 된다.

```python
plt.imshow(img_hopper)
```

**주의:** 위 그림에서 세로가 329개의 칸으로, 가로가 500개의 칸으로 구분되어 있음을 확인할 수 있다.
